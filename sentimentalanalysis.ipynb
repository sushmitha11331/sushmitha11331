This project analyzes text (like tweets or reviews) and classifies them as **Positive**, **Negative**, or **Neutral** using Python and TextBlob.  


```python
!pip install textblob
from textblob import TextBlob


reviews = [
    "I love this phone! The camera quality is amazing ğŸ“±ğŸ˜",
    "This product is terrible and broke after one use ğŸ˜¡",
    "The movie was okay, not great but not bad either ğŸ¬",
    "Absolutely fantastic experience! Highly recommend ğŸ‘",
    "Worst service ever. Totally disappointed ğŸ˜ ",
    "Itâ€™s an average product, could be better."
]


print("ğŸ”¹ Analyzing sample reviews...\n")
for review in reviews:
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity  # range: -1 (negative) to +1 (positive)

    if polarity > 0:
        sentiment = "Positive ğŸ™‚"
    elif polarity < 0:
        sentiment = "Negative ğŸ™"
    else:
        sentiment = "Neutral ğŸ˜"

    print(f"ğŸ“ Review: {review}")
    print(f"â¡ï¸ Sentiment: {sentiment}\n")


print("ğŸ”¹ Now try your own sentence below ğŸ‘‡")
text = input("Enter your review or tweet: ")
score = TextBlob(text).sentiment.polarity

if score > 0:
    print("ğŸ™‚ Positive")
elif score < 0:
    print("ğŸ™ Negative")
else:
    print("ğŸ˜ Neutral")

```
