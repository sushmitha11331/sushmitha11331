# üì® Task 1 ‚Äì Spam Email Classifier  
This project detects whether a text message is **spam** or **ham** (not spam) using Python and scikit-learn.  




```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score


url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
data = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])
print("üìä Sample Data:\n", data.head())


X_train, X_test, y_train, y_test = train_test_split(
    data['message'], data['label'], test_size=0.2, random_state=42)


vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_tfidf, y_train)


y_pred = model.predict(X_test_tfidf)
print("\n‚úÖ Accuracy:", accuracy_score(y_test, y_pred))


sample = [
    "You have won a free iPhone! Click here to claim.",
    "Hey, are we meeting tomorrow?"
]
sample_tfidf = vectorizer.transform(sample)
print("\nüîç Predictions:", model.predict(sample_tfidf))
```
